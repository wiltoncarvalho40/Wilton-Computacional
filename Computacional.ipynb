```python
import matplotlib.pyplot as plt

```

```python
h = 0.01
x_i = -1
x_f = 2
intervalo_1_2 = []

for i in range(int((x_f - x_i)/h)):
  intervalo_1_2.append(x_i + i*h)

fxi = []
ci_fxi = []

for i in range (len(intervalo_1_2)):
  x = intervalo_1_2[i]
  f_x_i = x**4+2*x
  ci_fxi.append(f_x_i*h)
  fxi.append(f_x_i)



```

```python
delta_x = 0.01
x = 2
derivada = (((x+delta_x)**4 + 2*(x+delta_x)) - ((x-delta_x)**4 + 2*(x-delta_x)))/(2*delta_x)
derivada_real = 4*x**3 + 2

print(f"Derivada: {derivada:.6f}")
print(f"Derivada Real: {derivada_real:.6f}")
```

```python
delta_x = 0.01
x = 2
derivada = ((12*(x+delta_x)**2) - (12*(x-delta_x)**2))/(2*delta_x)
derivada_real = 12*x**2

print(f"Derivada: {derivada:.6f}")
print(f"Derivada Real: {derivada_real:.6f}")

import matplotlib.pyplot as plt
plt.plot(x, derivada, 'ro')

plt.title('Integração Numérica - Regra da Extremidade Esquerda')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.legend()
#if Salvar_png == True:
# plt.savefig(Diretorio + 'integracao_extremidade_esquerda.png', dpi = dpi)
#elif Salvar_pdf == True:
# plt.savefig(Diretorio + 'integracao_extremidade_esquerda.pdf')
plt.show()
```

```python
import numpy as np

delta_x = np.pi/10
x = np.pi/4
f_linha_x = (np.sin(x+delta_x) - np.sin(x-delta_x))/(2*delta_x)
f_linha_x_analitico = np.cos(x)

print(f"Derivada numérica: {f_linha_x:.6f}")
print(f"Derivada Real: {f_linha_x_analitico:.6f}")
```
