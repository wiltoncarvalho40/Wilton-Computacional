```python
import numpy as np
from numpy.linalg import norm

# Novos valores para os vetores
v1 = np.array([5, 0])
v2 = np.array([0, 5])


norma_v1 = norm(v1, 2)
norma_v2 = norm(v2, 2)


produto_escalar = np.dot(v1,v2)


angulo_v1_v2 = np.acos(produto_escalar/(norma_v1*norma_v2))*(180/np.pi)


print(f'Vetor v1: {v1}')
print(f'Vetor v2: {v2}')
print(f'Norma de v1: {norma_v1}')
print(f'Produto Escalar: {produto_escalar}')
print(f'Ângulo entre v1 e v2 (graus): {angulo_v1_v2}')
```

```python
a = 0.5
b = -2
c = 2
x = np.linspace(-10, 10, 100)
z = np.linspace(-1, 2, 100)
X = np.array([x, z, np.ones(len(x))]).T
y = a * X.T[0] + b*X.T[1] + c*X.T[2]

X_transposto = X.T
X_trans_X = np.dot(X_transposto, X)
X_inverso = np.linalg.inv(X_trans_X)
x_trans_y = np.dot(X_transposto, y)


theta = np.dot(X_inverso, x_trans_y)

print(theta)
```
